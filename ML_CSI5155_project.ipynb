#### Helper function
Used to print out and compare the different models

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score, precision_score, recall_score, f1_score

"""
This function calculates and prints the accuracy, precision, recall, and F1 score
based on the predicted and actual values. It also plots the confusion matrix.

Parameters:
- y_test (array-like): The actual values.
- y_pred (array-like): The predicted values.

Returns:
None
"""
def test_model_quality(y_test, y_pred):

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    print("Accuracy:", accuracy)

    # Calculate precision
    precision = precision_score(y_test, y_pred)
    print("Precision:", precision)

    # Calculate recall
    recall = recall_score(y_test, y_pred)
    print("Recall:", recall)

    # Calculate F1 score
    f1 = f1_score(y_test, y_pred)
    print("F1 Score:", f1)

    # Plot confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.show()
    
```

### Data Splitting

```python
def split_data(X, y, unlabeled_size):
    
    # Split the dataset into labeled and "unlabeled" (we actually just hide the labels for demonstration)
    X_labeled, X_unlabeled, y_labeled, y_unlabeled_actual = train_test_split(X, y, test_size=unlabeled_size, random_state=42)
    
    # Simulate the unlabeled dataset by setting labels to -1
    y_unlabeled = [-1 for _ in range(len(X_unlabeled))]

    return X_labeled, X_unlabeled, y_labeled, y_unlabeled, y_unlabeled_actual
```

### Self-training Algorithm

```python
from sklearn.semi_supervised import SelfTrainingClassifier

def run_self_training_model(X_labeled, X_unlabeled, y_labeled, y_unlabeled, X_test, y_test):
    # Create a self-training algorithm model using SVC, adding default threshold and criterion to show what is used.
    self_training_model = SelfTrainingClassifier(svm.SVC(probability=True, random_state=42), threshold=0.75, criterion='threshold')

    # Combine the labeled and unlabeled datasets
    X_combined_self_training = pd.concat([X_labeled, X_unlabeled])
    y_combined_self_training = pd.concat([pd.Series(y_labeled), pd.Series(y_unlabeled)])

    # Fit the model on the labeled and unlabeled data
    self_training_model.fit(X_combined_self_training, y_combined_self_training)

    # Predict the labels for the test data
    y_pred_self_training = self_training_model.predict(X_test)

    # Evaluate the model
    test_model_quality(y_test, y_pred_self_training)
```

### Co-training Algorithm

```python

# TODO There is a mistake in the for loop, doesn't properly swap output
def run_co_training(X_labeled, X_unlabeled, y_labeled, X_test, y_test, confidence_threshold=0.8, num_iterations=1):
    # Simulate two different views by splitting the features
    features_set_1 = X_labeled.columns[:len(X_labeled.columns) // 2]
    features_set_2 = X_labeled.columns[len(X_labeled.columns) // 2:]

    X_view1_labeled_co_training = X_labeled[features_set_1]
    X_view1_unlabeled_co_training = X_unlabeled[features_set_1]
    X_view2_labeled_co_training = X_labeled[features_set_2]
    X_view2_unlabeled_co_training = X_unlabeled[features_set_2]

    # Initialize classifiers
    svc_classifier_co_training_1 = svm.SVC(probability=True, random_state=42)
    svc_classifier_co_training_2 = svm.SVC(probability=True, random_state=43) # Different random state for diversity

    # Co-training loop with confidence thresholding
    for _ in range(num_iterations):
        # Train classifiers on the initially labeled data
        svc_classifier_co_training_1.fit(X_view1_labeled_co_training, y_labeled)
        svc_classifier_co_training_2.fit(X_view2_labeled_co_training, y_labeled)

        # Generate pseudo-labels for unlabeled data in the opposite view
        pseudo_probs_view1 = svc_classifier_co_training_2.predict_proba(X_view1_unlabeled_co_training)
        pseudo_probs_view2 = svc_classifier_co_training_1.predict_proba(X_view2_unlabeled_co_training)

        # Identify high-confidence predictions
        high_confidence_view1 = np.max(pseudo_probs_view1, axis=1) > confidence_threshold
        high_confidence_view2 = np.max(pseudo_probs_view2, axis=1) > confidence_threshold

        # Select high-confidence instances and their pseudo-labels
        high_conf_X_view1 = X_view1_unlabeled_co_training[high_confidence_view1]
        high_conf_labels_view1 = np.argmax(pseudo_probs_view1[high_confidence_view1], axis=1)

        high_conf_X_view2 = X_view2_unlabeled_co_training[high_confidence_view2]
        high_conf_labels_view2 = np.argmax(pseudo_probs_view2[high_confidence_view2], axis=1)

        # Combine the high-confidence pseudo-labeled data with the originally labeled data
        X_combined_view1_co_training = np.vstack((X_view1_labeled_co_training, high_conf_X_view1))
        X_combined_view2_co_training = np.vstack((X_view2_labeled_co_training, high_conf_X_view2))
        y_combined_view1_co_training = np.concatenate((y_labeled, high_conf_labels_view1))  # Assuming equal contribution for simplicity

        # Update the labeled datasets for the next iteration
        X_view1_labeled_co_training = X_combined_view1_co_training
        X_view2_labeled_co_training = X_combined_view2_co_training
        y_labeled = y_combined_view1_co_training

    # Retrain classifiers on the combined sets for final evaluation
    svc_classifier_co_training_1.fit(X_combined_view1_co_training, y_combined_view1_co_training)
    svc_classifier_co_training_2.fit(X_combined_view2_co_training, y_combined_view1_co_training)

    # Evaluate the classifiers on the test set
    y_pred_view1_co_training = svc_classifier_co_training_1.predict(X_test[features_set_1])
    y_pred_view2_co_training = svc_classifier_co_training_2.predict(X_test[features_set_2])

    # Evaluate the model
    test_model_quality(y_test[features_set_1], y_pred_view1_co_training)
    test_model_quality(y_test[features_set_2], y_pred_view2_co_training)
```

### Semi-supervisable ensemble

```python
# TODO this function hasn't been verified yet
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
def run_semi_supervised_ensemble(X_labeled, y_labeled, X_unlabeled, X_test, y_test, confidence_threshold=0.9):
    # Train an initial model on the labeled data
    initial_model_ensemble = svm.SVC(probability=True, random_state=42)
    initial_model_ensemble.fit(X_labeled, y_labeled)

    # Predict on the unlabeled data
    y_unlabeled_pred = initial_model_ensemble.predict(X_unlabeled)
    y_unlabeled_proba = initial_model_ensemble.predict_proba(X_unlabeled)

    # Select high-confidence predictions to add to the labeled set
    confidence_threshold = 0.9  # This threshold can be adjusted
    high_confidence_indices = np.max(y_unlabeled_proba, axis=1) > confidence_threshold
    X_high_confidence = X_unlabeled[high_confidence_indices]
    y_high_confidence = y_unlabeled_pred[high_confidence_indices]

    # Create an augmented labeled dataset by combining the original labeled data with high-confidence predictions
    X_augmented = pd.concat([X_labeled, X_high_confidence])
    y_augmented = pd.concat([y_labeled, pd.Series(y_high_confidence)])

    # Train new models on the augmented dataset
    model1 = RandomForestClassifier(n_estimators=100, random_state=42)
    model2 = svm.SVC(probability=True, random_state=42)

    # Create an ensemble of models
    ensemble = VotingClassifier(estimators=[
        ('rf', model1),
        ('svc', model2)
    ], voting='soft')

    # Train the ensemble on the augmented dataset
    ensemble.fit(X_augmented, y_augmented)

    # Evaluate the ensemble model on a held-out validation set (or the original test set)
    y_pred = ensemble.predict(X_test)
    test_model_quality(y_test, y_pred)
```

### An intrinsically semi-supervised learning method 
For this example, we use label propagation as an example of intrinsically semi-supervised learning technique

```python
# TODO This function hasn't been verified yet
from sklearn.semi_supervised import LabelPropagation
def run_label_propagation(X_labeled, X_unlabeled, y_labeled, y_unlabeled, X_test, y_test):
    # Create a label propagation model
    label_prop_model = LabelPropagation()

    # Combine the labeled and unlabeled datasets
    X_combined_label_prop = pd.concat([X_labeled, X_unlabeled])
    y_combined_label_prop = pd.concat([pd.Series(y_labeled), pd.Series(y_unlabeled)])

    # Fit the model on the combined dataset
    label_prop_model.fit(X_combined_label_prop, y_combined_label_prop)

    # Predict the labels for the test data
    y_pred_label_prop = label_prop_model.predict(X_test)

    # Evaluate the model
    test_model_quality(y_test, y_pred_label_prop)
```
